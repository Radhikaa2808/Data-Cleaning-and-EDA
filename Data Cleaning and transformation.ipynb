# Data Cleaning
 - correct the data
- Duplicate value
- capitalise
- Noise values and unwanted spaces
- data type correct
-  Null values

```python
df.Quantity.unique()
```

```python
df.Quantity.replace("five",5,inplace=True) #this is older way of writing code
df["Quantity"] = df["Quantity"].replace("five",5) #this is new way of writing code
```

```python
df.drop_duplicates(inplace=True)
#it will drop all the rows which are repeated not by just checking single column.
```

```python
df.duplicated().sum()
```

```python
cat = [] we created an empty list to store the names of categorical columns.
for i in df.columns:
    if df[i].dtypes=="O":
        cat.append(i)
    else:
        pass
```

```python
cat
```

```python
cat.remove('Quantity')
```

```python
cat
```

```python
for i in cat:
    df[i] = df[i].str.capitalize()
```

```python
new_col = ['Payment_Method','Product_Category']
for i in new_col:
    df[i] = df[i].replace(r'[^a-zA-Z0-9]',regex=True)
```

```python
df["Product_ID"] = df["Product_ID"].replace(r'[^a-zA-Z0-9]',regex=True)
```

```python
df["Order_ID"] = df["Order_ID"].str.strip().replace(" ","")
```

```python
df[["Price","Quantity"]] = df[["Price","Quantity"]].astype("float32")
```

```python
df["Order_Date"]=pd.to_datetime(df["Order_Date"])
```

```python
df.dtypes
```

```python
df.Price.skew()
```

```python
df[["Price","Quantity"]] = df[["Price","Quantity"]].fillna(df[["Price","Quantity"]].mean())
```

```python
df.isnull().sum()
```

```python
cat
```

```python
cat.remove('Product_ID')
```

```python
for i in cat:
    df[i] = df[i].fillna("Unknown")
```

```python
df["Product_ID"] = df["Product_ID"].fillna( df["Product_ID"].mode()[0]) #most repeated value in the column.
```

```python
# Data Transformation | creating new columns (feature engineering)
df["Revenue"] = df["Price"]*df["Quantity"]
```

```python
df["Price_Category"] = df["Price"].apply(new_col) #feature engineering
```
