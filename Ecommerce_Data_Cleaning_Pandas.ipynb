# E-Commerce Data Cleaning using Pandas


```python
import pandas as pd
import numpy as np
```

## Load Dataset

```python
df = pd.read_csv(r'ecommerce_dataset.csv')
df.head()
```

## Initial Data Understanding

```python
df.info()
df.isnull().sum()
```

```python
for i in df.columns:
    print(i,df[i].sort_values().unique(),"\n",sep="\n")
```

## Basic Cleaning

```python
df['Customer_Name'] = df['Customer_Name'].fillna('Guest')
df = df.drop_duplicates()
df['Unit_Price'] = df['Unit_Price'].astype(str).str.replace(r'[^0-9.]', '', regex=True).astype(float)
df.Unit_Price
```

## Quantity Cleaning

```python
word_to_num = {'one':1,'two':2,'three':3}
df['Quantity'] = df['Quantity'].replace(word_to_num)
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df = df[df['Quantity'] > 0]
```

```python
df['Product_Category'] = df['Product_Category'].replace({
    'electronic': 'electronics',
    'elec': 'electronics'
})
```

## Date Standardization

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
```

## Advanced Cleaning

```python
df['Order_ID'] = df['Order_ID'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
df.dtypes
```

## Feature Engineering

```python
df['Total_Amount'] = df['Unit_Price'] * df['Quantity']
df['Order_Year'] = df['Order_Date'].dt.year
```

## Regex-based Cleaning

```python
df['Customer_Name'] = df['Customer_Name'].str.replace(r'[^A-Za-z ]','', regex=True)
```

## Final Cleaned Data

```python
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df[["Unit_Price","Quantity"]] = df[["Unit_Price","Quantity"]].fillna(df[["Unit_Price","Quantity"]].mean())
```

```python
def new_col(data):
    if data<=150:
        return "Low Price of Product"
    elif data>150 and data<=370:
        return "Medium Price  Of Product"
    else:
        return "High Cost Of Product"


df["Price_Category"] = df["Unit_Price"].apply(new_col)
```
